```python
from collections import defaultdict

from earthquake_prediction.utils import read_test
from earthquake_prediction.experiments import cnn_encoder_lstm

import torch
import pandas as pd
```

```python
exp = cnn_encoder_lstm.Experiment('lstm2', pd.DataFrame([]))
exp.load()
```

```python
test_files = read_test()
```

```python
results = defaultdict(list)

for name, inp in test_files:
    pred = exp.predict(torch.from_numpy(inp['acoustic_data'].values).unsqueeze(0).unsqueeze(0).float())
    results['seg_id'].append(name)
    results['time_to_failure'].append(pred)
    
```

```python
res = pd.DataFrame(results)
# len(res[res['time_to_failure'] < 3])
res
```

```python
res.to_csv('./data/submission.csv', index=False)
```
